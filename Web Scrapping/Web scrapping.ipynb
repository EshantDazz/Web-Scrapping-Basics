**Function to get any link and download in our system**

```python
def download(link,name):
    r=requests.get(link)
    s=bb(r.text,'html.parser')
    f=open(name,'wb')
    f.write(s.prettify('utf=8'))
    f.close()
```

```python
link=input("Enter the link")
name=input("Enter the name of the file with .html extension please")
download(link,name)
```
