```python
import requests
from bs4 import BeautifulSoup as bb
r=requests.get("https://www.geeksforgeeks.org/")
s=bb(r.text,'html.parser')
```

```python
ss=s.find('div',id='main')
```

```python
ss.prettify
```

```python
sss=ss.find('div',class_='article--container')
```

```python
sss.text
```

# Article basic info

```python
s.find('div',id='main')
```

```python
s.find('div',id='main').find_all('span')
```

List of all spans

```python
print(s.find('div',id='main').find_all('div'))
```

```python
s.find('div',id='main').find_all('a href')
```

```python
for i in s.find('div',id='main').find_all('span'):
    print(i.text)
```

```python
s.find('div',class_='header-main__wrapper')
```

```python
 s.find('div',class_='header-main__wrapper').text
```

```python
for i in s.find_all('div',class_='header-main__wrapper'):
    print(i.text)
```

```python
for i in s.find_all('div',class_='header-main__wrapper')[:]:
    print(i.text)
#We can do slicing on this as well
```
