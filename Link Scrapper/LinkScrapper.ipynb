**ALL the links present**

```python
links = []

for i in str(s).split('<a'):
    if ('href' in i):
        links.append(str(i).split('</a>')[0].split('href="')[1].split('"')[0])
```

```python
links
```

```python
fd = open('links.txt','a')

for link in links:
    fd.write(link + '\n')
    
fd.close()
```
